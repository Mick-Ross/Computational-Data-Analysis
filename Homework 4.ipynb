```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

## Problem 0: Setting up fake background data

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

## Problem 1: Subsets of data

1a) Create a scatter plot of the first 1000 points.

```python
plt.title(f'Random Data', fontsize='large')
plt.xlabel('Time',fontsize='large')
plt.ylabel('Magnitude',fontsize='large')
plt.scatter(t[:1000], bkgd[:1000])
```

1b) Generalize your code so that you can make a plot of any X contiguous points.

```python
def contiguousPt(start, numPoints, t, data, step = None):
    if ((start + numPoints) > len(data)):
        return "Out of range"
    
    plt.title(f'Random Data', fontsize='large')
    plt.xlabel('Time',fontsize='large')
    plt.ylabel('Magnitude',fontsize='large')
    plt.scatter(t[start:start+numPoints:step], bkgd[start:start+numPoints:step])
    
```

```python
contiguousPt(50, 100, t, bkgd)
```

1c) Plotting every 100th data point.

```python
contiguousPt(0, len(t) - 1, t, bkgd, 100)
```

## Problem 2: Density

2a) Make a 2D histogram plot with a useful number of bins.

```python
plt.hist2d(t, bkgd, bins = 30)
plt.colorbar()

plt.title(f'Random Data', fontsize='xx-large')
plt.xlabel('Time',fontsize='large')
plt.ylabel('Magnitude',fontsize='large');
```

2b) Explain what is plotted in your plot.

We are plotting the density of points within each bin. We chose a number of bins which determined the dimensions of our square plot (numBins x numBins). The color of each bin determines the number of points within that given bin.

## Problem 3: Folding

3a) Make a new time variable that folds at a period of $2\pi$. Then make a scatter plot of all your data.

```python
time2 = t % [2*np.pi]
contiguousPt(0, len(t) - 1, time2, bkgd, 100)
plt.title(f'Random Data', fontsize='xx-large')
plt.xlabel('Time (mod $2\pi$)',fontsize='large');
```

3b) Make a 2D histogram plot of your folded data.

```python
plt.hist2d(time2, bkgd, bins = 30, cmap = 'magma')
plt.colorbar()

plt.title(f'Random Data', fontsize='xx-large')
plt.xlabel('Time (mod $2\pi$)',fontsize='large')
plt.ylabel('Magnitude',fontsize='large');
```
